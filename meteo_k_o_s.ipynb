```python
from bs4 import BeautifulSoup
import requests
import os
import zipfile
import codecs
import pandas as pd
import csv
```

```python
if not os.path.exists('./meteo_1'):
    os.mkdir('./meteo_1')

if not os.path.exists('./meteo_1/daily_meteo_1_extracted_zips'):
    os.mkdir('./meteo_1/daily_meteo_1_extracted_zips')

k_o_s = ['klimat', 'opad', 'synop']

for elem in k_o_s:
    if not os.path.exists(f'./meteo_1/daily_meteo_1_extracted_zips/{elem}'):
        os.mkdir(f'./meteo_1/daily_meteo_1_extracted_zips/{elem}')

k_d = []
o_d = []
s_d = []

daily_urls = [k_d, o_d, s_d]
```

```python
for i in range(2018, 2024):
    k_d.append(f'https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteo_1rologiczne/dobowe/klimat/{i}/')
    o_d.append(f'https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteo_1rologiczne/dobowe/opad/{i}/')
    s_d.append(f'https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteo_1rologiczne/dobowe/synop/{i}/')

for tab in daily_urls:
    for elem in tab:

        html_page = requests.get(elem)
        soup = BeautifulSoup(html_page.content, 'html.parser')

        for link in soup.find_all('a'):
            if link.get('href').endswith('.zip'):
                
                zip_file = link.get('href')
                file_url = elem + zip_file
                
                if elem in k_d:
                    download_path = os.path.join('./meteo_1/daily_meteo_1_extracted_zips/klimat', zip_file)
                elif elem in o_d:
                    download_path = os.path.join('./meteo_1/daily_meteo_1_extracted_zips/opad', zip_file)
                elif elem in s_d:
                    download_path = os.path.join('./meteo_1/daily_meteo_1_extracted_zips/synop', zip_file)

                urllib.request.urlretrieve(file_url, download_path)

                with zipfile.ZipFile(download_path, 'r') as zip_ref:
                    if elem in k_d:
                        zip_ref.extractall('./meteo_1/daily_meteo_1_extracted_zips/klimat')

                    elif elem in s_d:
                        zip_ref.extractall('./meteo_1/daily_meteo_1_extracted_zips/synop')

                    elif elem in o_d:
                        zip_ref.extractall('./meteo_1/daily_meteo_1_extracted_zips/opad')

                os.remove(download_path)
```

```python
directory_k = './meteo_1/daily_meteo_1_extracted_zips/klimat'

for file in directory_k:

    file_path = os.path.join(directory_k, file)

    with codecs.open(file_path, 'r', encoding='Windows-1250') as f:
        contents = f.read()

    with codecs.open(file_path, 'w', encoding='utf-8') as f:
        f.write(contents)

    df = pd.read_csv(file_path, encoding='utf-8')

    df.columns = df.columns.str.strip()
    df = df.dropna()
    df = df.drop_duplicates()
```
