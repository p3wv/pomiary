```python
paths = './meteo/daily_meteo_extracted_zips'
path = os.listdir(paths)

if not os.path.exists('./meteo/daily_meteo_csvs_cleaned/'):
    os.mkdir('./meteo/daily_meteo_csvs_cleaned/')

for p in path:
    if not os.path.exists(f'./meteo/monthly_meteo_csvs_cleaned/{p}/'):
        dest_dir = f'./meteo/daily_meteo_csvs_cleaned/{p}/'
    for file_name in os.listdir(p):
        file_path = os.path.join(p, file_name)
        if os.path.isfile(file_path):
            shutil.copy(file_path, dest_dir)

```
